```python
#https://web.stanford.edu/class/archive/cs/cs109.1166/problem12.html
url = "https://web.stanford.edu/class/archive/cs/cs109.1166/stuff/titanic.csv"
```

```python
import pandas as pd

#import titanic dataset
df = pd.read_csv(url)
df.head()
```

```python
df.describe(include="all")
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# Drop empty data 
df = df.dropna()
```

```python
df["Sex"]
```

```python
# fill empty data
 df{"Age"} = df{"Age"}.fillna(df{"Age"}.mode())
```

```python
#forward fill
df.fillna(method='ffill', inplace=True)

#backward fill
df.fillna(method='bfill', inplace=True)
```

```python
df.sort_values("Age", inplace=True)
```

```python
df.head()
```

```python
df = df.reset_index()
df.head()
```

```python
df.drop(columns=['index','Pclass'], inplace=True)
df.head()
```

```python
df["Relatives"] = df['Siblings/Spouses Aboard']+df['Parents/Children Aboard']
df['new_fare'] = df['Fare']/10
df.head()
```

```python
df_group = df.groupby('Sex')[['Age', 'Relatives']].mean()
df_group                    
```

```python
df['Fare'].median()
```

```python
#merging datasets
df_new = pd.read_csv(url)
df_new.head()
```

```python
df
```

```python
merged_df = pd.merge(df, df_new, on='Name', suffixes=('_df', 'df_new'))
merged_df.head()
```

```python
#label encoding
from sklearn.preprocessing import labelEncoder

encoder = labelEncoder()

df['Sex_encode'] = encoder.fit_transform(df['Sex'])
df.head()
```

```python
#0ne-hot encoding
df = pd.get_dummies(df, columns=['Sex'], drop_first=True)
df.head()
```
